```python
from turtle import color
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python
data=pd.read_csv("marketing.csv",sep="\t")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data1=data.dropna()
data1["Dt_Customer"] = pd.to_datetime(data1["Dt_Customer"],dayfirst=True)
data1.head()
```

```python
data1['Age']=2022-data1['Year_Birth']
data1
```

```python
data1["Expenditure"] = data["MntWines"]+ data["MntFruits"]+ data["MntMeatProducts"]+ data["MntFishProducts"]+ data["MntSweetProducts"]+ data["MntGoldProds"]
data1["Family_mem"]=data["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})
data1["Children"]=data["Kidhome"]+data["Teenhome"]
data1.head()
```

```python
from tkinter.tix import INTEGER
data1["Family_Size"] = data1["Family_mem"].replace({"Alone": 1, "Partner":2}) + data1["Children"]
data1["Is_Parent"] = np.where(data1['Children']> 0, 1, 0)
data1=data1.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
drop = ["Marital_Status", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
data2 = data1.drop(drop, axis=1)
data2['Income'] = data2['Income'].astype(int)
data2.info()
data2.head()
```

```python
import seaborn as sns
Plot = [ "Income", "Recency", "Dt_Customer", "Age", "Expenditure", "Is_Parent"]
plt.figure()
sns.pairplot(data2[Plot], hue= "Is_Parent",palette= (["#682F2F","#F3AB60"]))
plt.show()
```

```python
data2=data2[(data2['Age']<90)]
data2=data2[(data2['Income']<600000)]
plt.figure()
sns.pairplot(data2[Plot], hue= "Is_Parent",palette= (["#682F2F","#F3AB60"]))
plt.show()
```

```python
data2.head()
```

```python
dummies=['Education','Family_mem']
dummy_data=pd.get_dummies(data2[dummies])
data3=pd.concat([data2,dummy_data],axis=1)
data3.drop(dummies,axis=1,inplace=True)
data3.info()
```

```python
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response','Dt_Customer']
data3 = data3.drop(cols_del, axis=1)
```

```python

data4=pd.DataFrame(data3)
data4.shape
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=1,random_state=3)
cluster=model.fit_predict(data4.values)
model.cluster_centers_.shape
```

```python
Plot = [ "Income", "Recency", "Age", "Expenditure", "Is_Parent"]
plt.scatter(data4[Plot[1]],y=data4['Income'] ,s=20, cmap='winter')
plt.scatter(data4[Plot[2]],y=data4['Income'] ,s=20, cmap='summer')
plt.scatter(data4[Plot[3]],y=data4['Income'] ,s=20, cmap='prism')
plt.scatter(data4[Plot[4]],y=data4['Income'] ,s=20, cmap='gnuplot')
centers = model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='blue', s=10, alpha=0.9)
plt.show()
```
